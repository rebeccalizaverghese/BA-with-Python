```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('/content/RELIANCE.NS.csv')
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
x
```

```python
y
```

```python
y = y.reshape(len(y),1)
```

```python
from sklearn.svm import SVR
regressor = SVR()
regressor.fit(x, y)
```

```python
y_pred = regressor.predict(x)
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```

```python
regressor = SVR(kernel = 'rbf')
regressor.fit(x, y)
y_pred = regressor.predict(x)
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```

```python
regressor = SVR(kernel = 'linear')
regressor.fit(x, y)
y_pred = regressor.predict(x)
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```

```python
regressor = SVR(kernel = 'poly')
regressor.fit(x, y)
y_pred = regressor.predict(x)
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```

```python
regressor = SVR(kernel = 'rbf' ,C=1.0, epsilon=0.2)
regressor.fit(x, y)
y_pred = regressor.predict(x)
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```

```python
regressor = SVR(kernel = 'rbf',C=2, epsilon=0.2)
regressor.fit(x, y)
y_pred = regressor.predict(x)
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```

```python
regressor = SVR(kernel = 'rbf',C=.1, epsilon=0.2)
regressor.fit(x, y)
y_pred = regressor.predict(x)
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```

```python
regressor = SVR(kernel = 'rbf', degree=3, gamma='scale', coef0=0.0, tol=0.001, C=1.0, epsilon=0.1, shrinking=True, cache_size=200, verbose=False, max_iter=- 1)
regressor.fit(x, y)
y_pred = regressor.predict(x)
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```
